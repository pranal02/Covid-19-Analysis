```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('worldometer_coronavirus_daily_data.csv')
df.head()
```

```python
#converting the data columns to datetime type
df['date']=pd.to_datetime(df['date'])
```

```python
# Aggregate data as needed 
df_global = df.groupby('date').sum().reset_index()
```

```python
df_global.head()
```

```python
# visualize the data with Plotly

# Line plot for global confirmed. active cases and deaths over time
fig = px.line(df_global,x='date',y=['daily_new_cases','active_cases','cumulative_total_deaths'], title="Global COVID-19 Trends", labels={'value': 'Cases', 'variable': 'Metric'})
fig.show()
```

```python
# Bar chart for new cases per day
df_global['New Cases']=df_global['active_cases'].diff().fillna(0)
fig = px.bar(df_global,x='date',y='active_cases' ,title='Daily New covid 19 cases Globally' , labels={'New Cases':'New Cases per Day'},color_discrete_sequence=['red'])
fig.show()
```

```python

df_global=df_global.drop(columns=['New Cases'])
df_global.head()
```

```python
# Scatter Plot for Death Rate vs Active cases
# calculate Death rate and daily new cases
df_global['Death Rate']=df_global['daily_new_deaths']/df_global['daily_new_cases']
fig=px.scatter(df_global,x='Death Rate' ,y='daily_new_cases', title="Death Rate vs Recovery new cases")
fig.show()
```

```python
#Choropleth Map for country wise distribution
df_country_summary = df_global.groupby('country')[['daily_new_cases', 'active_cases', 'daily_new_deaths']].sum().reset_index()

# If only 'Country' column exists with multiple rows per country, count occurrences (example metric)


# Create a choropleth map for the given data
fig = px.choropleth(df_country_summary, 
                    locations='country', 
                    locationmode='country names',
                    color='active_cases',  # Assuming 'Count' as a metric
                    hover_name='country',
                    color_continuous_scale='Reds',
                    title="Global COVID-19 Cases by Country (Example Metric)")
fig.show()

```

```python
# Step 1: Create 'Month-Year' column for heatmap
df_global['Month-Year'] = df_global['date'].dt.to_period('M').astype(str)

# Step 2: Group by 'Month-Year' and aggregate the required columns
df_monthly = df_global.groupby('Month-Year')[['daily_new_cases', 'active_cases', 'daily_new_deaths']].sum().reset_index()

# Step 3: Transpose the data for the heatmap (Metrics as rows, Months as columns)
df_monthly_transposed = df_monthly.set_index('Month-Year').T

# Step 4: Create a heatmap with transposed data
fig = px.imshow(df_monthly_transposed,
                labels=dict(x="Month-Year", y="Metric", color="Count"),
                title="Heatmap of Monthly Case Counts")

fig.show()
```

```python
# Create a Time Series Model using Facebook Prophet
```

```python
df_prophet = df_global[['date', 'active_cases']]
df_prophet.columns = ['ds', 'y']
```

```python
# Initialize and fit the Prophet model
model = Prophet(interval_width=0.95)
model.fit(df_prophet)
```

```python
# Create a dataframe to hold future dates
future = model.make_future_dataframe(periods=7)  # 7 days into the future

# Predict the future
forecast = model.predict(future)

# View the predictions
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(7)
```

```python
# Creating combined plot
actual_trace = go.Scatter(
    x=df_prophet['ds'],
    y=df_prophet['y'],
    mode='lines+markers',
    name='Actual'
)

forecast_trace = go.Scatter(
    x=forecast['ds'],
    y=forecast['yhat'],
    mode='lines',
    name='Forecast'
)

upper_bound = go.Scatter(
    x=forecast['ds'],
    y=forecast['yhat_upper'],
    mode='lines',
    fill='tonexty',
    name='Upper Bound',
    line={'color': 'rgba(10,20,30,0)'}
)

lower_bound = go.Scatter(
    x=forecast['ds'],
    y=forecast['yhat_lower'],
    mode='lines',
    fill='tonexty',
    name='Lower Bound',
    line={'color': 'rgba(0,0,0,0)'}
)

fig = go.Figure(data=[actual_trace, forecast_trace, lower_bound, upper_bound])
fig.update_layout(title="COVID-19 Cases Forecast vs Actual", xaxis_title="Date", yaxis_title="Cases")
fig.show()
```
